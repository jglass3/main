# IMPORT LIBRARIES

```python
#Method used for visualing high demensional data, not for classification, so no train/testing
import pandas as pd                # For dataframe
import matplotlib.pyplot as plt    # For plotting data
import seaborn as sns              # For plotting data
from sklearn.manifold import TSNE  # For tSNE
```

# LOAD AND PREPARE DATA

Load the complete dataset (as opposed to the training or testing data) from the CSV file in the data directory. Separate the data matrix from the class variable. Name the data matrix as `df` and the column with class labels as `y`.

```python
# Imports the complete dataset
df = pd.read_csv('optdigits.csv')

# Separates the attributes P0-P63 into X
X = df.filter(regex='\d')

# Separates the class variable into y
y = df.y
```

```python
# Shows the first few rows of the data
df.head()
#P values tell how many pixels are actived within a given region, y = number
```

# t-SNE: MODEL DATA

This phase applies t-SNE to the training data with various values of the perplexity parameter and displays embeddings of data.

```python
# Sets up the t-SNE object with 2 components
#Have to specific number of components
tsne = TSNE(
    n_components=2,
    random_state=1)
```

```python
# Displays the t-SNE parameters
tsne.get_params()
```

## PERPLEXITY = 1

```python
# Sets up t-SNE with perplexity (parameter) = 1
#perplexity - how it combines high dimensionality data down into our 2 dimensional system
#Starting perplexity with 1, which you don't ever want to do
tsne = TSNE(
    n_components=2,
    perplexity=1,
    random_state=1)

# Transforms the attribute data
X_tf = tsne.fit_transform(X)

# Creates a scatterplot of the data embedding
sns.scatterplot(
    x=X_tf[:, 0],
    y=X_tf[:, 1],
    style=y,
    hue=y,
    palette=['red', 'green', 'blue'])
#Doesn't help us out
```

## PERPLEXITY = 2

```python
# Sets up t-SNE with perplexity = 2
tsne = TSNE(
    n_components=2,
    perplexity=2,
    random_state=1)

# Transforms the attribute data
X_tf = tsne.fit_transform(X)

# Creates a scatterplot of the data embedding
sns.scatterplot(
    x=X_tf[:, 0],
    y=X_tf[:, 1],
    style=y,
    hue=y,
    palette=['red', 'green', 'blue'])
```
