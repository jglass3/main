```python
#Association Rule Mining - prcoess that deploys pattern recognition to identify and quantify relationshiops between differentj
    #yet related items
#Popular use case: product placment optimatiztion at stores
#3 ways to measure association
    #Lift - Metrics that measures how much more often the antecedent and consequent occur together rather
        #than them occuring independently. 
        #Score>1: A is highly associated with C meaning if A is purchased will likely be purchased. 
        #Score<1: If A is purchased, C will not be purchased
        #Score=1: No association
    #Support - Relative frequency of an item within a dataset (ratio=variable total/total volume)
    #Confidence - Probability of seeing the consequent item ("then" term)within data, given that the data also
        #contains the antecednet("if" term). How likely an item is going to be purchased, if another
        #item is purchased
#Apriori is the algo that you use to implement association rule mining over data
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

# Data Format

```python
address = 'groceries.csv'
data = pd.read_csv(address)
```

```python
data.head()
```

# Data Coversion

```python
#data needs to be converted to sparse format so algo can work
#implementing OneHotEncoding
basket_sets = pd.get_dummies(data)
```

```python
basket_sets.head()
```

# Support Calculation

```python
#support tells how popular an item is based on all transactions
apriori(basket_sets, min_support=0.02)
```

```python
#generate new columns to include names of items
apriori(basket_sets, min_support=0.02, use_colnames=True)
```

```python
#single popular items only displayed, decrease the support to open up the list
df = basket_sets

frequent_itemsets = apriori(basket_sets, min_support=0.002, use_colnames=True)
#add length to see number of items sold. Ex 2 items sold - length = 2
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
#lot of combinations are being seen, filter out combos to only 3 or more
frequent_itemsets[frequent_itemsets['length'] >= 3]
```

# Association Rules

## Confidence

```python
#generate association rules with two inputs (mlx library)
#1.if the metric should be confidence or lift
#2. min level for these metrics
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules.head()
#minimum confidence level starts at .5, which is how likely C is purchased if A is purchased
#0 row - sauage is purchased, it's likely that frankfurter is purchased (confi = 1)
```

## Lift

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
#Lift - how more often item C and A occur in same transaction
#Greatest combo of being bought together is beef and citrus fruit (row 1 has slightly greather confi)
    #It is likely that if citrus fruit (A) is purchase beef will also be purchased (C)
```

## Lift and Confidence

```python
#filter out rules for optimal lift and confidence
#rules from data analyzed above
#
rules[(rules['lift'] >= 5) & (rules['confidence']>= 0.5)]
```
